3. Вычислите ориентированную площадь исходного прямоугольника. В каком порядке
нужно перечислить вершины прямоугольника, чтобы ориентированная площадь имела положительный знак?

```python
x0, y0 = -1, -0.5
width, height = 2, 1
vertices = np.array([[x0, y0],[x0+width, y0],[x0+width, y0+height],[x0, y0+height],[x0, y0]])
def oriented_area(vertices):
    n=len(vertices)
    area=0
    for i in range(n):
        x1, y1 = vertices[i]
        x2, y2 = vertices[(i+1)%n]
        area+=x1*y2-x2*y1
    return area/2
area = oriented_area(vertices)
print(area) #нужно перечислять вершины против часовой стрелки
```

```python
x0, y0 = -1, -0.5
width, height = 2, 1
vertices = np.array([[x0, y0],[x0+width, y0],[x0+width, y0+height],[x0, y0+height],[x0, y0]])
def oriented_area(vertices):
    n=len(vertices)
    area=0
    for i in range(n):
        x1, y1 = vertices[i]
        x2, y2 = vertices[(i+1)%n]
        area+=x1*y2-x2*y1
    return area/2
M=np.array([[-1,1],[1,1]])
trans_vertices=vertices @ M.T
new_area=oriented_area(trans_vertices)
det = np.linalg.det(M)
print(f"определитель: {det}")
print(f"о-я площадь: {new_area}") #знак изменился (как и абсолютная величина) из-за определителя
```
