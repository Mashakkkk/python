```python
p = np.array([[-0.5,0,-0.5],[0.5,0,-0.5],[0.5,0,0.5],[-0.5,0,0.5],[0,0.71,0]])
faces = [[0,1,4], [1,2,4], [2,3,4], [3,0,4]]
vect = np.array([[1,0,0],[0,1,0],[0,0,1]])
def oblique(f, alpha, sign=-1):
    a = np.radians(alpha)
    ca, sa = np.cos(a), np.sin(a)
    return np.array([[1, 0],[0, 1],[sign*f*ca, sign*f*sa]])#xyz из 3d в 2d
def rot_y(p, angle):
    theta = np.radians(angle)
    c, s = np.cos(theta), np.sin(theta)
    R = np.array([[c, 0, s], [0, 1, 0], [-s, 0, c]])
    return p @ R.T
def draw(alpha, angle, proj_type):
    if proj_type == 'Кавалье':
        f = 1
        title = 'Проекция Кавалье (f=1)'
    else:
        f = 0.5
        title = 'Проекция Кабине (f=0.5)'
    mat = oblique(f, alpha)
    p_rot = rot_y(p, angle)
    p2d = p_rot @ mat
    b2d = vect @ mat
    fig, ax = plt.subplots(figsize=(8,8))
    for face in faces:
        verts = p2d[face]
        poly = Polygon(verts, closed=True, facecolor='lightgray', edgecolor='black', linewidth=1.5, alpha=0.7)
        ax.add_patch(poly)
    for i, (x, y) in enumerate(p2d):
        ax.text(x, y, str(i), color='blue', fontsize=12, fontweight='bold', ha='center', va='bottom')
    colors = ['red', 'green', 'blue']
    labels = ['X', 'Y', 'Z']
    for i, vec in enumerate(b2d):
        ax.arrow(0, 0, vec[0], vec[1], color=colors[i], head_width=0.05, head_length=0.08, linewidth=2, alpha=0.7)
        ax.text(vec[0]*1.1, vec[1]*1.1, labels[i], color=colors[i], fontsize=12)
    ax.scatter(p2d[:,0], p2d[:,1], c='r', s=30)
    ax.set_title(f'{title}\nα={alpha}, поворот={angle}')
    ax.set_aspect('equal')
    ax.grid()
    ax.set_xlim(-2, 2.5)
    ax.set_ylim(-1.5, 2)
    plt.show()
interact(draw,alpha=IntSlider(min=0, max=360, step=5, value=45, description='α'),angle=IntSlider(min=0, max=360, step=5, value=0, description='Поворот'),proj_type=Dropdown(options=['Кавалье', 'Кабине'], value='Кавалье', description='Проекция'))
```

## Задание №2

1. Создайте шахматное поле из клеток.
2. На поле из клеток нарисуйте пирамиду.
3. Создайте ползунки для ее перемещения.

```python
def proj(p):
    f = 0.5
    alpha = 45 
    sign = -1
    a = np.radians(alpha)
    ca, sa = np.cos(a), np.sin(a)
    P = np.array([[1, 0, 0],[0, 1, 0],[sign * f * ca, sign * f * sa, 0]])
    return p @ P
sq = np.array([[0,0,0],[1,0,0],[1,0,1],[0,0,1]])
def draw(x, y, z):
    plt.figure(figsize=(8,6))
    for i in range(6):
        for j in range(6):
            s = sq + [i, 0, j]
            s2d = proj(s)
            color = 'red' if (i+j)%2 else 'blue'
            plt.fill(s2d[:,0], s2d[:,1], color, edgecolor='black', linewidth=0.5, alpha=0.7)
    pp = proj(p + [x, y, z])
    for face in faces:
        plt.fill(pp[face][:,0], pp[face][:,1], '#555555', edgecolor='black', linewidth=1, alpha=0.9)
    plt.axis('equal')
    plt.title(f'Косоугольная проекция (кабине)\nПирамида: x={x:.1f}, y={y:.1f}, z={z:.1f}')
    plt.show()
interact(draw, x=(-2,7,0.5), y=(0,2,0.1), z=(-2,7,0.5))
```
