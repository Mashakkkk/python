## Задание №1

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from ipywidgets import interact, IntSlider, FloatSlider
import ipywidgets as widgets
from matplotlib.patches import Circle
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
a=4
b=3
def angl(a1,a2):
    plt.figure(figsize=(6,6))
    x=y=0
    a_1=np.radians(a1)
    a_2=np.radians(a2)
    x1=a*np.cos(a_1)
    y1=a*np.sin(a_1)
    x2=x1+b*np.cos(a_1+a_2)
    y2=y1+b*np.sin(a_1+a_2)
    plt.plot([x,x1, x2], [y,y1, y2],marker='o',linestyle='-',color='black')
    plt.grid(True)
    plt.axhline(color='black')
    plt.axvline(color='black')
    print(f"p1: ({x1:.3f}, {y1:.3f})")
    print(f"p2: ({x2:.3f}, {y2:.3f})")
    plt.xlim(-1, 10)
    plt.ylim(-1, 10)
    plt.show()
interact(angl,a1=widgets.IntSlider(min=0, max=90, step=2),a2=widgets.IntSlider(min=0, max=180, step=2))
```

## Задание №2

```python
def lin(R, r, d, mode="epi", n=2000):
    t = np.linspace(0, 2*np.pi, n)
    if mode == "epi": #эпитрохоида
        x = (R + r)*np.cos(t) - d*np.cos((R + r)/r * t)
        y = (R + r)*np.sin(t) - d*np.sin((R + r)/r * t)
    elif mode == "hypo": #гипотрохоида
        x = (R - r)*np.cos(t) + d*np.cos((R - r)/r * t)
        y = (R - r)*np.sin(t) - d*np.sin((R - r)/r * t)
    plt.figure(figsize=(6,6))
    plt.plot(x, y, color="black")
    plt.axhline(color="black")
    plt.axvline(color="black")
    plt.grid(True)
    plt.axis("equal")
    plt.title(f"{mode}lin (R={R}, r={r}, d={d})")
    plt.show()
def draw(mode, R, r, d):
    lin(R, r, d, mode)
interact(draw,mode=widgets.Dropdown(options=["epi", "hypo"], value="epi"),R=widgets.IntSlider(min=1, max=20, step=1, value=10),r=widgets.IntSlider(min=1, max=10, step=1, value=3),d=widgets.IntSlider(min=0, max=20, step=1, value=5))
```

## Задание №3

```python
def ani(R, r, d, t_max):
    t = np.linspace(0, t_max, 500)
    x_center = (R - r) * np.cos(t)
    y_center = (R - r) * np.sin(t)
    phi = (R - r) / r * t #угол вращения круга
    x = x_center + d * np.cos(phi)
    y = y_center - d * np.sin(phi)
    x_c = (R - r) * np.cos(t_max) #текущий x центра
    y_c = (R - r) * np.sin(t_max)
    phi_c = (R - r) / r * t_max
    x_p = x_c + d * np.cos(phi_c) #текущий x точки
    y_p = y_c - d * np.sin(phi_c)
    circle_t = np.linspace(0, 2*np.pi, 200) #точки для окружности
    cx = x_c + r * np.cos(circle_t) #иксы мал круга
    cy = y_c + r * np.sin(circle_t)
    plt.figure(figsize=(6,6))
    big_x = R * np.cos(circle_t)
    big_y = R * np.sin(circle_t)
    plt.plot(big_x, big_y, color="gray", linestyle="--", label="большой круг")
    plt.plot(cx, cy, color="blue", alpha=0.5, label="малый круг")
    plt.plot(x, y, color="black")
    plt.plot(cx, cy, color="blue", alpha=0.5)
    plt.plot(x_p, y_p, 'ro')
    plt.plot(x_c, y_c, 'bo')
    plt.plot([x_c, x_p], [y_c, y_p], color="red")
    plt.axhline(color='black')
    plt.axvline(color='black')
    plt.axis("equal")
    plt.grid(True)
    plt.show()
interact(ani,R=widgets.IntSlider(min=2, max=20, step=1, value=10),r=widgets.IntSlider(min=1, max=10, step=1, value=3),d=widgets.IntSlider(min=0, max=15, step=1, value=5),t_max=widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1))
```

## Задание №4

```python
def curve(a1, k1, a2, k2, a3, k3, t_max):
    t = np.linspace(0, t_max, 2000)
    p1 = a1 * np.exp(1j * k1 * t)
    p2 = a2 * np.exp(1j * k2 * t)
    p3 = a3 * np.exp(1j * k3 * t)
    p = p1 + p2 + p3
    x = np.real(p)
    y = np.imag(p)
    tc = t_max
    P1 = a1 * np.exp(1j * k1 * tc)
    P2 = a2 * np.exp(1j * k2 * tc)
    P3 = a3 * np.exp(1j * k3 * tc)
    P = P1 + P2 + P3
    plt.figure(figsize=(6,6))
    plt.plot(x, y, color="black")
    plt.quiver(0, 0, np.real(P1), np.imag(P1), angles='xy', scale_units='xy', scale=1, color="red")
    plt.quiver(np.real(P1), np.imag(P1),np.real(P2), np.imag(P2), angles='xy', scale_units='xy', scale=1, color="blue")
    plt.quiver(np.real(P1+P2), np.imag(P1+P2),np.real(P3), np.imag(P3), angles='xy', scale_units='xy', scale=1, color="green")
    plt.quiver(0, 0, np.real(P), np.imag(P),angles='xy', scale_units='xy', scale=1, color="black", width=0.005)
    plt.axhline(color='gray')
    plt.axvline(color='gray')
    plt.axis("equal")
    plt.grid(True)
    plt.show()
interact(curve,a1=widgets.FloatSlider(min=0, max=5, step=0.1, value=1),k1=widgets.FloatSlider(min=-10, max=10, step=0.1, value=1),a2=widgets.FloatSlider(min=0, max=5, step=0.1, value=1),k2=widgets.FloatSlider(min=-10, max=10, step=0.1, value=2),a3=widgets.FloatSlider(min=0, max=5, step=0.1, value=1),k3=widgets.FloatSlider(min=-10, max=10, step=0.1, value=3),t_max=widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1))
```
